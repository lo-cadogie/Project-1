```python
#dependencies
import pandas as pd
import numpy as np

```

```python
#import file
spotify_file = "Resources\Project1_spotify_data.csv"
first_spotify_df = pd.read_csv(spotify_file)
first_spotify_df.head()
```

```python
# remove null values
first_spotify_df = first_spotify_df.dropna()

## Remove duplicate songs if week matches
spotify_df = first_spotify_df.drop_duplicates(subset=["track_name", "rank", "week",], keep='first')
spotify_df
```

```python
#create a copy df for merging with a new genre
spotify_df2 = spotify_df.copy()
```

```python
#Read in the genre categories file
genre_file = "Resources\genres.csv"
genre_df = pd.read_csv(genre_file)
genre_df.head()
```

```python
#combine the current df with a genre category df
spotify_df3 = pd.merge(spotify_df2, genre_df, on = 'artist_genres', how ='left')
spotify_df3
```

```python
#doing some tests
spotify_df3['genre_category'].value_counts()
```

```python
#doing some tests
spotify_df3['track_name'].value_counts()
```

```python
# output the clean dataset

spotify_df3.to_csv('Resources/spotify_clean.csv', index = False)
```
