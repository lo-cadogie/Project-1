```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("Resources\spotify_clean.csv")
df.head()
```

```python
genre_count = df.groupby(["genre_category"]).count()["track_name"]
plot = genre_count.plot.pie(y='', figsize=(15,15), labels = ["Dance", "Hip-Hop", "International", "Other", "Pop", "Rock"], title = "Genre Shares", autopct='%1.1f%%', colors = ["royalblue", "peru", "green", "maroon", "darkviolet", "saddlebrown"])
plt.ylabel("")
plt.show()
```
